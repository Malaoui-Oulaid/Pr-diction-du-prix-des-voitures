# ***Exploration et nettoyage des données***

```python
data['CompanyName'] = data['CarName'].apply(lambda x: x.split(' ')[0])
```

```python
def replace_values(data, a, b):
    data['CompanyName'].replace({a: b}, inplace=True)
```

```python
data['CompanyName'].unique()
```

```python
replace_values(data, 'maxda','mazda')
replace_values(data, 'porcshce','porsche')
replace_values(data, 'toyouta','toyota')
replace_values(data, 'vokswagen','volkswagen')
replace_values(data, 'vw','volkswagen')
```

```python
data['CompanyName'].value_counts()
```

```python
data.drop(['car_ID','CarName','symboling'],axis=1, inplace=True)
```

```python
#visualiser le variable cible
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Boxplot
sns.boxplot(y=data['price'], ax=axes[0], color='#FF5733')
axes[0].set_title('distribution de price')
axes[0].set_ylabel('Price')

# Histogramme
sns.histplot(data['price'], ax=axes[1], kde=True, color='#33FF57', bins=30)
axes[1].set_title('distribution de price')
axes[1].set_xlabel('Price')
axes[1].set_ylabel('Fréquence')

plt.tight_layout()
plt.show()
```

```python
def categorical_visualisation(col):
    plt.figure(figsize=(16, 6))

    plt.subplot(1,3,1)
    sns.countplot(data=data, x=col, order=data[col].value_counts().index, palette='pastel')
    plt.title(f"Nombre de {col}", pad=20, fontweight='black', fontsize=10)
    plt.xticks(rotation=45)

    plt.subplot(1,3,2)
    sns.boxplot(data=data, x=col, y='price', palette='pastel')
    plt.title(f"Distribution des prix par type de {col}", pad=20, fontweight='black', fontsize=10)
    plt.xticks(rotation=45)

    plt.subplot(1,3,3)
    df2=pd.DataFrame(data.groupby(col)['price'].mean().sort_values(ascending=False))
    sns.barplot(data=df2, x=df2.index, y='price', palette='pastel')
    plt.title(f"{col} vs Average Price", pad=20, fontweight='black', fontsize=10)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
def numerical_col_visualization(col):
    plt.figure(figsize=(15, 6))
    a = 1
    for c in [col]:
        plt.subplot(1,3,a)
        sns.regplot(data=data, y='price', x=c, scatter_kws={'color': 'red'}, line_kws={'color': 'green'}, marker='p')
        plt.title(f"{c} vs Price", pad=10, fontweight=12, fontsize=10)
        a=a+1

    plt.tight_layout()
    plt.show()
```

```python
categorical_columns = data.select_dtypes('object').columns.tolist()
numerical_columns = data.select_dtypes(['number', 'int', 'float']).columns.tolist()
```

```python
for col in categorical_columns:
    categorical_visualisation(col)
    print(200*"*")
```

```python
for col in numerical_columns:
    numerical_col_visualization(col)
    print(200*"*")
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(data[numerical_columns].corr(), annot=True, cmap='viridis', fmt=".2f")
plt.title('Matrice de corrélation des variables numériques', fontsize=16)
plt.show()
```
